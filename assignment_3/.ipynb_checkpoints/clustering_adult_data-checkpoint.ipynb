```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
url_test = 'http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test'
```

```python
# load  the training data from 
train_data = pd.read_csv(url,
                         header=None,
                         names=['age', 'workclass','fnlwgt','education','education_num','maritalstatus',
                                'occupation','relationship','race','sex','capital_gain','capital_loss',
                                'hours_per_week','native_country', 'income'],
                         na_values = ' ?')
train_data.head()
```

```python
# load  the training data from 
test_data = pd.read_csv(url_test,
                         header=None,
                         names=['age', 'workclass','fnlwgt','education','education_num','maritalstatus',
                                'occupation','relationship','race','sex','capital_gain','capital_loss',
                                'hours_per_week','native_country', 'income'],
                         na_values = ' ?')
test_data.head()
```

## 2. Data Preprocessing ##

```python
def preprocess_data(data):
    ''' Change the categorical columns to numeric '''
    data = data.dropna()
    data = data.drop_duplicates()
    for column in data.columns:
        if data[column].dtype == object:
            data[column] = data[column].astype('category').cat.codes
    return data
    
```

```python
train_df = preprocess_data(train_data)
test_df = preprocess_data(test_data)
```

```python
print(train_df.shape)
train_df.head()
```

```python
print(test_df.shape)
test_df.head()
```

```python
# generating features and target variable for training and testing
X = train_df.drop(columns=['income'], axis=0)
y = train_df.income

X_test = test_df.drop(columns='income',axis=0)
y_test = test_df.income
```

# 4. Experiments and Results 

## Deep Neural Network

```python
from keras.models import Sequential, save_model
from keras.callbacks import EarlyStopping
from keras.layers import Dense
from keras.utils import to_categorical
from keras.optimizers import SGD, adam
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# early_stopping_monitor = EarlyStopping(patience=2)
input_shape = (X.shape[1],)

# # Create the new model: model_2
# model_1 = Sequential()

# # Add the hidden layers
# model_1.add(Dense(50, activation='relu', input_shape=input_shape))
# model_1.add(Dense(50, activation='relu'))
# # model_1.add(Dense(50, activation='relu'))


# # Add the output layer
# model_1.add(Dense(2, activation='sigmoid'))

```

### Effect of learning rate in optimization

```python
%%time
subs = 150
plt.figure(figsize=(25, 6))
for lr in [0.04, 0.05, 0.06, 0.07, 0.08]:
    # Create the new model: model_2
    model_1 = Sequential()

    # Add the hidden layers
    model_1.add(Dense(50, activation='relu', input_shape=input_shape))
    model_1.add(Dense(50, activation='relu'))
    # model_1.add(Dense(50, activation='relu'))


    # Add the output layer
    model_1.add(Dense(2, activation='sigmoid'))
    
    subs = subs+1
    # Compile model_2
    model_1.compile(optimizer=SGD(lr=lr), loss='binary_crossentropy', metrics=['accuracy'])
    # Fit model_1
    model_1_training = model_1.fit(X_scaled,
                                 to_categorical(y),
                                 epochs=60,
                                 validation_data=(X_test_scaled, to_categorical(y_test)),
                                 verbose=False)
    # Create the plot
    plt.subplot(subs)
    plt.plot(model_1_training.history['loss'], color='b', label='Validation Error')
    plt.plot(model_1_training.history['val_loss'], color='r',label='Training Error')
    plt.xlabel('Epochs')
    plt.ylabel('Log_loss score')
    plt.legend(loc='best')
    plt.title('Neural Network Learning Curve(lr = {})'.format(lr))
plt.show()
```

### Best Model

```python
from keras.callbacks import TensorBoard
from time import time
tensorboard = TensorBoard(log_dir="logs/{}".format(time()))

# Create the new model: model_2
model_tuned = Sequential()

# Add the hidden layers
model_tuned.add(Dense(50, activation='relu', input_shape=input_shape))
model_tuned.add(Dense(50, activation='relu'))
model_tuned.add(Dense(50, activation='relu'))

# Add the output layer
model_tuned.add(Dense(2, activation='sigmoid'))
# Compile model
model_tuned.compile(optimizer=SGD(lr =0.06), loss='binary_crossentropy', metrics=['accuracy'])

# Fit model
model_training = model_tuned.fit(X_scaled, to_categorical(y), epochs=60, validation_split=0.3, verbose=False, callbacks=[tensorboard])
loss, acc = model_tuned.evaluate(X_test_scaled, to_categorical(y_test), verbose=False)
```

```python
print("Test Accuracy:", acc)
print("Train Accuracy:", max(model_training.history['acc']))
print("Loss Score:", loss)
```
