### Neural Network Weight Optimization

#### Random Hill Climbing

```python
nn_rhc = pd.read_csv('RHC/RHC_LOG.csv')
nn_rhc.head()
```

```python
def plot_error(data, title):
    plt.figure(figsize=(8,4))
    plt.plot(data.iteration, data.MSE_trg, label='Training')
    plt.plot(data.iteration, data.MSE_tst, label='Testing')
    plt.xlabel('Iteration')
    plt.ylabel('Mean Squared Error')
    plt.title('Error vs Iteration plot for '+title)
    plt.show

def plot_acc(data, title):
    plt.figure(figsize=(8,4))
    plt.plot(data.iteration, data.acc_trg, label='Training')
    plt.plot(data.iteration, data.acc_tst, label='Testing')
    plt.xlabel('Iteration')
    plt.ylabel('Accuracy')
    plt.title('Accuracy vs Iteration plot for '+title)
    plt.show
    
```

```python
plot_error(nn_rhc, 'Random Hill Climbing')
```
